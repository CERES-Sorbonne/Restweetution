```python
def build_skeletton(df):
    start_date = df['created_at'].iloc[0]
    end_date = df['created_at'].iloc[-1]
    nodes = "Id;Label"
    nodes_dic = {}
    edges = "Source;Target;Weight"
    default_weight = 100
    index_days = 0
    index_sha1 = 0
    prev_id = None 
    delta = timedelta(days=1)
    while start_date <= end_date:
        # create day id
        id_current = f"d{index_days}"
        index_days += 1
        # add day to the nodes
        nodes += f"\n{id_current};{start_date.strftime('%Y-%m-%d')}"
        # create a link between current day and previous day if there is one
        if prev_id:
            edges += f"\n{prev_id};{id_current};{default_weight}"
        # get all sha1 at this date and group them by sha1
        images_at_this_date = df.loc[df['created_at'] == start_date].groupby('sha1', as_index=False).count()
        if not images_at_this_date.empty:
            # for each sha1 at this date
            for index, line in images_at_this_date.iterrows():
                # create an id only if the node does not already exist
                sha1 = line['sha1']
                if sha1 in nodes_dic:
                    id_sha1 = nodes_dic[sha1]
                else:
                    id_sha1 = f"s{index_sha1}"
                    nodes += f"\n{id_sha1};{sha1}"
                    index_sha1 += 1
                # add an edge between the day and the sha1 with weight = nb of occurences that day
                edges += f"\n{id_current};{id_sha1};{line['created_at']}"
        # increment date                             
        start_date += delta
        # save current_id
        prev_id = id_current
    return nodes, edges
```

```python
nodes, edges = build_skeletton(df)
with open(os.path.join(path, 'nodes.csv'), 'w') as f:
    f.write(nodes)
with open(os.path.join(path, 'edges.csv'), 'w') as f:
    f.write(edges)
```
